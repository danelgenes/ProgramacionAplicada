```python

import pygame 
import math


negro = (0, 0, 0);
blanco = (255, 255, 255);
azul = (0, 0, 255);
largo_ladrillo = 23;
alto_ladrillo = 15;
 
class Bloque(pygame.sprite.Sprite):
    
 
    def __init__(self, color, x, y):
        
        super().__init__()
         
        self.image = pygame.Surface([largo_ladrillo, alto_ladrillo]);
        self.image.fill(color);
        self.rect = self.image.get_rect();
        
        self.rect.x = x;
        self.rect.y = y;
 
 
class Pelota(pygame.sprite.Sprite):
     

    velocidad = 12;
    x = 0.0;
    y = 180.0;
     
    rumbo = 200;
 
    largo = 10;
    alto = 10;
     
    def __init__(self):
        
        super().__init__()
         
        
        self.image = pygame.Surface([self.largo, self.alto]);
         
       
        self.image.fill(blanco);
         
       
        self.rect = self.image.get_rect();
         
        self.alto_screen = pygame.display.get_surface().get_height();
        self.largo_screen = pygame.display.get_surface().get_width();
     
    def botar(self, diff):
         
        self.rumbo = (180 - self.rumbo) % 360;
        self.rumbo -= diff;
     
    def update(self):
        
        rumbo_radianes = math.radians(self.rumbo);
         
        self.x += self.velocidad * math.sin(rumbo_radianes);
        self.y -= self.velocidad * math.cos(rumbo_radianes);
         
  
        self.rect.x = self.x;
        self.rect.y = self.y;
         
        
        if self.y <= 0:
            self.botar(0);
            self.y = 1;
             
        
        if self.x <= 0:
            self.rumbo = (360 - self.rumbo) % 360
            self.x = 1
             
        
        if self.x > self.largo_screen - self.largo:
            self.rumbo = (360 - self.rumbo) % 360
            self.x = self.largo_screen - self.largo - 1
         
        
        if self.y > 600:
            return True
        else:
            return False
 
class Protagonista(pygame.sprite.Sprite):
   
     
    def __init__(self):
        
        super().__init__()
         
        self.largo = 75
        self.alto = 15
        self.image = pygame.Surface([self.largo, self.alto])
        self.image.fill((blanco))
         
       
        self.rect = self.image.get_rect()
        self.alto_screen = pygame.display.get_surface().get_height()
        self.largo_screen = pygame.display.get_surface().get_width()
 
        self.rect.x = 0
        self.rect.y = self.alto_pantalla-self.alto
     
    def update(self):
        #posicion actualizada
        pos = pygame.mouse.get_pos()
        # movimiento barra delimitado por mouse
        self.rect.x = pos[0]
       
        if self.rect.x > self.largo_screen - self.largo:
            self.rect.x = self.largo_pantalla - self.largo
 

pygame.init()
 

pantalla = pygame.display.set_mode([800, 600])
 

pygame.display.set_caption('Breakout')
 

pygame.mouse.set_visible(0)
 

fuente = pygame.font.Font(None, 36)
 

fondo_pantalla = pygame.Surface(pantalla.get_size())
 

bloques = pygame.sprite.Group()
pelotas = pygame.sprite.Group()
todos_los_sprites = pygame.sprite.Group()
 

protagonista = Protagonista()
todos_los_sprites.add(protagonista)
 

pelota = Pelota()
todos_los_sprites.add(pelota)
pelotas.add(pelota)
 

top = 80
 

numero_de_bloques = 32
 
for fila in range(5):
   
    for columna in range(0, numero_de_bloques):
        
        bloque = Bloque(AZUL, columna * (largo_screen + 2) + 1, top)
        bloques.add(bloque)
        todos_los_sprites.add(bloque)
    
    top += alto_bloque + 2
 

reloj = pygame.time.Clock()
 

game_over = False
 

salir_programa = False
 

while not salir_programa:
 
    
    reloj.tick(30)
 
    
    pantalla.fill(NEGRO)
     
    
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            salir_programa = True
     
   
    if not game_over:
        
        protagonista.update()
        game_over = pelota.update()
 
    if game_over:
        texto = fuente.render("Game Over", True, BLANCO)
        textopos = texto.get_rect(centerx=fondo_pantalla.get_width()/2)
        textopos.top = 300
        pantalla.blit(texto, textopos)
     
    
    if pygame.sprite.spritecollide(protagonista, pelotas, False):
       
        diff = (protagonista.rect.x + protagonista.largo/2) - (pelota.rect.x+pelota.largo/2)
         
       
        pelota.rect.y = pantalla.get_height() - protagonista.rect.alto - pelota.rect.alto - 1
        pelota.botar(diff)
     
    
    bloquesmuertos = pygame.sprite.spritecollide(pelota, bloques, True)
     
    
    if len(bloquesmuertos) > 0:
        pelota.botar(0)
         
        
            game_over = True
     
    todos_los_sprites.draw(pantalla)
 
    
    pygame.display.flip()
 
pygame.quit()
```
